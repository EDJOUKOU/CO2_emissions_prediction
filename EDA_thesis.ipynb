### Conclusion of the descriptive statistics

1. The carbon emissions in most sectors exhibit their mean values, which are greater than their median, so they are **all positively skewed except the emissions in the forestry, farm, waste, and other sectors**, which are left-skewed, as shown in the table above. However, the degree of skewness varies a bit among the sectors. Emissions data in power, residential, farm, waste, and other sectors are fairly symmetrical; the biomass and forestry sectors are moderately skewed, while the transport and industry sectors are highly skewed.
2. The box and whisker plot confirms what is discussed above. In addition, among all sectors, the median lines of the other and, especially the forestry sectors, lie outside the boxes of other boxplots, suggesting a major difference. There is also a **noticeable variability in the interquartile range** of each sector, which explains differences in emissions levels.
3. **Outliers** are visible only within the transport and industry sectors.
4. The excess kurtosis is everywhere less than zero, suggesting that **the distribution has a platykurtic curve**. However, the kurtosis values in the transport and industry sectors are above zero, which demonstrates that the distribution of these sectors has a leptokurtic curve.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV  # Changed to RidgeCV for auto-tuning
from sklearn.model_selection import cross_val_score
from SALib.sample import saltelli
from SALib.analyze import sobol
from statsmodels.stats.outliers_influence import variance_inflation_factor


# Now use these new columns
features = ['gdp_per_cap', 'pop_density', 'arable_land', 'e_power', 'forest']
target = 'others'

df_clean = df[features + [target]].dropna()

X = df_clean[features]
y = df_clean[target]
```

```python
# Step 2: Standardize features and check initial VIF
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Pre-regularization VIF check
print("\n--- Pre-regularization VIF ---")
vif_pre = pd.DataFrame()
vif_pre["Variable"] = X.columns
vif_pre["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
print(vif_pre.sort_values("VIF", ascending=False))
```

```python
# Step 3: Ridge Regression with automatic alpha tuning
alphas = np.logspace(-3, 3, 50)  # Test wide range of regularization strengths
ridge = RidgeCV(alphas=alphas, cv=5, scoring='neg_mean_squared_error')
ridge.fit(X_scaled, y)

# Cross-validation results
print(f"\nOptimal alpha: {ridge.alpha_:.4f}")
print(f"Cross-validated R²: {np.mean(cross_val_score(ridge, X_scaled, y, cv=5)):.3f}")
```

```python
# Step 4: Post-regularization diagnostics
# Method 1: Calculate VIF on original features (shows Ridge doesn't technically change VIF)
# Ridge reduces coefficient variance but doesn't eliminate multicollinearity in the data
vif_post = pd.DataFrame()
vif_post["Variable"] = X.columns
vif_post["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
print("VIF on original features (note Ridge doesn't change input data structure):")
print(vif_post.sort_values("VIF", ascending=False))

# Method 2: Better approach - show how Ridge shrinks coefficients despite high VIF
print("\n--- Coefficient Shrinkage Report ---")
coef_report = pd.DataFrame({
    'Variable': X.columns,
    'Coefficient': ridge.coef_,
    'Abs_Coefficient': np.abs(ridge.coef_),
    'VIF': vif_post["VIF"]
}).sort_values('Abs_Coefficient', ascending=False)

print(coef_report.round(4))
```

```python
# Step 5: Sobol Global Sensitivity Analysis (Updated for SALib 1.5+)
from SALib.sample import sobol as sobol_sample  # Updated import

problem = {
    'num_vars': X_scaled.shape[1],
    'names': X.columns.tolist(),
    'bounds': [[-3, 3]] * X_scaled.shape[1]  # Standardized range
}

# Generate samples with power-of-2 for better convergence
N = 1024  # Using 2^10 for proper Sobol sequence properties
param_values = sobol_sample.sample(problem, N, calc_second_order=True)

# Predict using the ridge model
Y_sobol = ridge.predict(param_values)

# Sobol analysis
sobol_result = sobol.analyze(problem, Y_sobol, print_to_console=False)

# Step 6: Enhanced Results presentation
sensitivity_df = pd.DataFrame({
    'Variable': problem['names'],
    'First-order (S1)': sobol_result['S1'],
    'Total-order (ST)': sobol_result['ST'],
    'Ridge_Coefficient': ridge.coef_,
    'Abs_Coefficient': np.abs(ridge.coef_),
    'VIF': vif_post["VIF"].values  # From previous VIF calculation
}).sort_values(by='Total-order (ST)', ascending=False)

# Calculate confidence intervals (if available)
if 'S1_conf' in sobol_result:
    sensitivity_df['S1_Confidence'] = sobol_result['S1_conf']
if 'ST_conf' in sobol_result:
    sensitivity_df['ST_Confidence'] = sobol_result['ST_conf']

print("\n--- Combined Results with Diagnostics ---")
print(sensitivity_df.round(4))
```

```python
# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(121)
sns.barplot(x='Ridge_Coefficient', y='Variable', 
            data=sensitivity_df.sort_values('Ridge_Coefficient', key=abs, ascending=False))
plt.title("Ridge Coefficients\n(alpha={:.2f})".format(ridge.alpha_))

plt.subplot(122)
sns.scatterplot(x='Ridge_Coefficient', y='Total-order (ST)', 
                hue='Variable', data=sensitivity_df, s=100)
plt.axvline(0, color='grey', linestyle='--')
plt.title("Sensitivity vs Coefficient Strength (others)")

plt.tight_layout()
plt.savefig("C:\\Users\\User\\Desktop\\Tech Industry\\Data Science\\Thesis\\Pictures\\ED12.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Select only the emission sectors and year column
emission_sectors = ['power', 'transport', 'residential', 'industry', 'biomass',
       'forestry', 'farm', 'wastes', 'others']
emission_data = df[['year'] + emission_sectors].set_index('year')

print("Original Emission Data (first 5 years):")
print(emission_data.head())
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Step 1: Normalize the data across sectors
scaler = StandardScaler()
normalized_data = scaler.fit_transform(emission_data)

# Convert back to dataframe for easier handling
normalized_df = pd.DataFrame(normalized_data, 
                            index=emission_data.index, 
                            columns=emission_data.columns)

print("\nNormalized Emission Data (first 5 years):")
print(normalized_df.head())
```

```python
# Step 2: Perform PCA for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(normalized_df.T)  # Transpose to cluster sectors

# Create a dataframe for the principal components
pca_df = pd.DataFrame(data=principal_components, 
                      columns=['PC1', 'PC2'],
                      index=emission_data.columns)

print("\nExplained variance ratio:", pca.explained_variance_ratio_)
```

```python
# Step 3: Cluster the sectors based on emissions
# First, determine optimal number of clusters using the elbow method
inertia = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_df.T)  # Transpose to cluster sectors
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 6), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.savefig("C:\\Users\\User\\Desktop\\Tech Industry\\Data Science\\Thesis\\Pictures\\ED13.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Based on the elbow plot, the slope flattens noticeably after k=2 
#  let's consider k = 2
optimal_clusters = 2
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(normalized_df.T)

# Add cluster labels to the PCA dataframe
pca_df['Cluster'] = clusters

# Step 4: Label each sector by emission level cluster
# Create cluster labels based on the mean emissions
cluster_means = normalized_df.T.groupby(clusters).mean().mean(axis=1)
cluster_labels = {cluster: label for cluster, label in 
                 zip(cluster_means.sort_values().index, 
                     ['low', 'high'])}

pca_df['Emission Level'] = pca_df['Cluster'].map(cluster_labels)

print("\nCluster Assignments:")
print(pca_df[['Emission Level']])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Emission Level', 
               palette={'low': 'green', 'high': 'red'},
               s=100)
for sector in pca_df.index:
    plt.text(pca_df.loc[sector, 'PC1']+0.05, pca_df.loc[sector, 'PC2'], sector,
             rotation=-6, ha='left', va='bottom')  # Added rotation and alignment
plt.title('Sector Clustering by Emission Patterns')
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.grid()
plt.savefig("C:\\Users\\User\\Desktop\\Tech Industry\\Data Science\\Thesis\\Pictures\\ED14.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Reshape the data for trend analysis
melted_df = emission_data.reset_index().melt(id_vars='year', 
                                            value_vars=emission_sectors,
                                            var_name='sector', 
                                            value_name='emissions')

# Add cluster labels to the melted dataframe
melted_df['Emission Level'] = melted_df['sector'].map(pca_df['Emission Level'])

# Plot emission trends by cluster
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_df, x='year', y='emissions', hue='Emission Level', 
            style='sector', palette={'low': 'green', 'high': 'red'},
            linewidth=2)
plt.title('Emission Trends by Sector and Cluster (1970-2023)')
plt.xlabel('Year')
plt.ylabel('Emissions (MtCO2eq')
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig("C:\\Users\\User\\Desktop\\Tech Industry\\Data Science\\Thesis\\Pictures\\ED15.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate some statistics about each cluster
cluster_stats = melted_df.groupby(['Emission Level', 'year'])['emissions'].mean().unstack('Emission Level')
cluster_stats = cluster_stats.describe().T

print("\nCluster Statistics:")
print(cluster_stats.round(2))
```

```python
# Additional analysis: Yearly patterns
# Calculate the mean emissions for each cluster by year
yearly_cluster_means = melted_df.groupby(['year', 'Emission Level'])['emissions'].mean().unstack()

# Plot yearly patterns
plt.figure(figsize=(10, 5))
yearly_cluster_means.plot(kind='line', 
                         color={'low': 'green', 'high': 'red'})
plt.title('Mean Yearly Emissions by Cluster')
plt.xlabel('Year')
plt.ylabel('Mean Emissions (MtCO2eq)')
plt.show()
```

```python
# Calculate the percentage change over time for each sector
melted_df['pct_change'] = melted_df.groupby('sector')['emissions'].pct_change() * 100

# Analyze volatility by cluster
volatility = melted_df.groupby(['Emission Level', 'sector'])['pct_change'].std().unstack('Emission Level')
print("\nEmission Volatility (Standard Deviation of Yearly Percentage Changes):")
print(volatility)
```
