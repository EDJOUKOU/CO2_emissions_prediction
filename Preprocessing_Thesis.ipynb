```python
# import necessary libraries

import numpy as np
import pandas as pd

# import the emissions data

df1 = pd.read_csv('emissions_data.csv')
df1.head()
```

```python
df1.tail()
```

```python
df1 = df1.iloc[0:54,:]
df1.tail()
```

```python
# let's check the null values
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
# fill in null values
df1 = df1.interpolate(method='linear', limit_direction='both')
```

```python
df1.isnull().sum()
```

```python
# the columns still having null values do not have any data. Let's fill them with zero

df1 = df1.fillna(0)
```

```python
df1.isnull().sum()
```

```python
# Create a dictionary mapping the new sectors to their constituent columns
sector_mapping = {
    'Power Industry': [
        'Carbon dioxide (CO2) emissions from Power Industry (Energy) (Mt CO2e)',
        'Methane (CH4) emissions from Power Industry (Energy) (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Power Industry (Energy) (Mt CO2e)'
    ],
    'Transport': [
        'Carbon dioxide (CO2) emissions from Transport (Energy) (Mt CO2e)',
        'Methane (CH4) emissions from Transport (Energy) (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Transport (Energy) (Mt CO2e)'
    ],
    'Residential': [
        'Carbon dioxide (CO2) emissions from Building (Energy) (Mt CO2e)',
        'Methane (CH4) emissions from Building (Energy) (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Building (Energy) (Mt CO2e)'
    ],
    'Industries': [
        'Carbon dioxide (CO2) emissions from Industrial Combustion (Energy) (Mt CO2e)',
        'Methane (CH4) emissions from Industrial Combustion (Energy) (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Industrial Combustion (Energy) (Mt CO2e)',
        'Carbon dioxide (CO2) emissions from Industrial Processes (Mt CO2e)',
        'Methane (CH4) emissions from Industrial Processes (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Industrial Processes (Mt CO2e)',
        'F-gases emissions from Industrial Processes (Mt CO2e)'
    ],
    'Solid Fuels and Biomass': [
        'solid fuels & emissions from biomass'
    ],
    'Forestry and Land Use': [

        'Carbon dioxide (CO2) net fluxes from LULUCF - Total excluding non-tropical fires (Mt CO2e)'
    ],
    'Agriculture': [
        'Carbon dioxide (CO2) emissions from Agriculture (Mt CO2e)',
        'Methane (CH4) emissions from Agriculture (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Agriculture (Mt CO2e)'
    ],
    'Wastes': [
        'Carbon dioxide (CO2) emissions from Waste (Mt CO2e)',
        'Methane (CH4) emissions from Waste (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Waste (Mt CO2e)'
    ],
    'Others': [
        'Carbon dioxide (CO2) emissions from Fugitive Emissions (Energy) (Mt CO2e)',
        'Methane (CH4) emissions from Fugitive Emissions (Energy) (Mt CO2e)',
        'Nitrous oxide (N2O) emissions from Fugitive Emissions (Energy) (Mt CO2e)',
        'Non specified'
    ]
}

# Create a new dataframe with year column
grouped_df = pd.DataFrame()
grouped_df['year'] = df1['year']

# Sum the columns for each sector
for sector, columns in sector_mapping.items():
    # Find which columns exist in the original dataframe
    existing_columns = [col for col in columns if col in df1.columns]
    if existing_columns:
        grouped_df[sector] = df1[existing_columns].sum(axis=1)
    else:
        grouped_df[sector] = 0


# Display the new dataframe
grouped_df.head()
```

```python
grouped_df.tail()
```

```python
# let's rename the column properly

grouped_df.rename(columns={'Power Industry':'power', 'Transport':'transport', 'Residential':'residential',
                  'Industries':'industry', 'Solid Fuels and Biomass':'biomass', 'Forestry and Land Use':'forestry',
                  'Agriculture':'farm', 'Wastes':'wastes', 'Others':'others'}, inplace=True)
grouped_df.head()
```

```python
# change the year column into a datetime object

grouped_df['year'] = pd.to_datetime(grouped_df['year'], format = '%Y')
```

```python
#check the data type of each column

grouped_df.dtypes
```

```python
# let's verify the dimensions of the data frame

grouped_df.shape
```

```python
grouped_df.columns
```

```python
# Let's add columns that computes the total of CO2 emissions including & exclusing LULUCF

sectors = ['power', 'transport', 'residential', 'industry', 'biomass',
       'forestry', 'farm', 'wastes', 'others']

sectors_L = ['power', 'transport', 'residential', 'industry', 'biomass',
        'farm', 'wastes', 'others']

grouped_df['gross_total'] = grouped_df[sectors].sum(axis=1)
grouped_df['exl_total'] = grouped_df[sectors_L].sum(axis=1)
```

```python
# let's rename the columns of the data frame df2

df2.rename(columns={'Carbon intensity of GDP (kg CO2e per constant 2015 US$ of GDP)':'carbon_intensity',
                   'GDP per capita (current US$)':'gdp_per_cap',
                   'Population density (people per sq. km of land area)':'pop_density', 
                   'Energy use (kg of oil equivalent per capita)':'energy_use',
                   'Arable land (hectares)':'arable_land',
                   'Electric power consumption (kWh per capita)':'e_power',
                   'Forest area (sq. km)':'forest'}, inplace=True)
```

```python
# merge the data frames grouped_df and df2 into a single one
df = pd.merge(grouped_df, df2, on='year', how='inner')

df.head() 
```

```python
# format the year column to exhibit only year

df['year'] = pd.to_datetime(df['year'], format = '%Y').dt.year

df.head()
```
